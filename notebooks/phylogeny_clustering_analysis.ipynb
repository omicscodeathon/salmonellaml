```python
df_training=pd.read_csv("../data/chisquare_training_features_final.csv")
df_training
```

```python
# decode the columns
custom_encoding_map = {0:'-', 1:'A', 2:'T', 3:'G', 4:'C'}
df_training=df_training.drop(columns=['Unnamed: 0','Isolation source'])
df_training
df_training['sra_accession_category_counrty'] = df_training['sra_accession'] + '_' + df_training['category'] + '_' + df_training['country']
```

```python
df_training=df_training.drop(columns=['category','sra_accession','BioSample','sra_accession','country'])
df_training
```

```python
# Get the name of the last column
last_column = df_training.columns[-1]

# Move the last column to the first position
df_training.insert(0, last_column, df_training.pop(last_column))
df_training
```

```python
df_training.replace(custom_encoding_map, inplace=True)
df_training
```

```python
# Set the first column as the index
df_training.set_index('sra_accession_category_counrty', inplace=True)
df_training
```

```python
# Define a function to concatenate the sequences
def concatenate_sequence(row):
    return ''.join(row)

# Apply the function to each row to concatenate sequences
df_training['sequence'] = df_training.apply(concatenate_sequence, axis=1)
df_training
```

```python
df_fasta=df_training[['sequence']]
df_fasta
# Save to a FASTA-like file
with open('../data/snps.fasta', 'w') as f:
    for index, row in df_fasta.iterrows():
        header = index
        sequence = ''.join(row.values)
        f.write(f'>{header}\n{sequence}\n')
```

```python
import pandas as pd
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

# Assuming df_trial contains the relevant encoded columns for clustering
encoded_columns = df_trial.columns[6:]  # Adjust the index based on your actual data

# Extract features for clustering
X = df_trial.head(20)[encoded_columns]

# Handle missing values (either impute or drop)
# X = X.dropna()  # Drop rows with missing values
X = X.fillna(0)  # Impute missing values (replace NaN with 0)

# Calculate linkage matrix using UPGMA
linkage_matrix = linkage(X, method='average')  # Do not transpose X to cluster rows

# Plot horizontal dendrogram with one label per cluster and category colors
plt.figure(figsize=(12, 6))

# The 'above_threshold_color' parameter ensures that clusters above the color threshold are displayed in grey
dendrogram_data = dendrogram(linkage_matrix, orientation='left', labels=None, leaf_rotation=0, leaf_font_size=8, above_threshold_color='grey')

# Get the category labels for the leaves
category_labels = df_trial['category'].reindex(X.index)

# Get unique categories and assign colors
unique_categories = df_trial['category'].unique()
category_colors = dict(zip(unique_categories, plt.cm.tab10.colors[:len(unique_categories)]))

# Map category labels to colors
colors = [category_colors.get(category, 'grey') for category in category_labels]

# Create a dictionary to map category names to colors for legend consistency
legend_category_colors = dict(zip(unique_categories, colors))

# Custom function to color branches based on category colors
def color_branches(node_color, leaf_color, ax):
    if leaf_color is not None:
        ax.axvline(node_color, color=leaf_color)

# Color branches based on category colors
for node_color, leaf_color in zip(dendrogram_data['icoord'], colors):
    color_branches(node_color[1], leaf_color, plt.gca())

# Add category colors as a legend using consistent colors
legend_handles = [Line2D([0], [0], marker='o', color='w', markerfacecolor=legend_category_colors[category], markersize=10, label=category)
                  for category in unique_categories]

plt.legend(handles=legend_handles, title='Categories', loc='upper right')

plt.title('UPGMA Dendrogram with Category Colors')
plt.xlabel('Distance')
plt.ylabel('Samples')
plt.show()
```
